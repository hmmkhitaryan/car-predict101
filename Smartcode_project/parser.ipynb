```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import datetime
import pandas as pd
import numpy as np
import re
```

```python
all_cat_list = []
```

```python
data = pd.read_excel(r'C:\Users\hamik\Desktop\Smartcode_project\auto_data.xlsx')
```

```python
start_time = time.time()
counter = 0
for n in range(20):
    # num = np.random.randint(3, 250)
    page_time = time.time()
    driver = webdriver.Chrome()
    # url = f'https://www.list.am/category/23/{120 - (n+1)}'
    url = f'https://www.list.am/category/23/{n+1}?bid=0&_a15=4%2C5'

    for k in range(96):
        element_time = time.time()
        driver.get(url)
    
        try:
            element = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, f"//*[@id='contentr']/div[4]/div[1]/a[{k+1}]"))
            )
            element.click()
            
        except Exception as e:
            print("Oops! An error has occured, but we got it!")
            element = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, f"//*[@id='contentr']/div[3]/div[1]/a[{k+1}]"))
            )
            element.click()
            
        clicked_url = element.get_attribute('href')
        driver.get(clicked_url)

        cat_list = []
        n_data = {}
        
        price = driver.find_elements(By.XPATH, "//*[@id='abar']/div[2]/span[1]")
        for pr in price:
            n_data['Price'] = pr.text
    
        n_data['Id'] = clicked_url[len(clicked_url)-8:]
    
        for j in range(3, 10, 2):
            attr_elements = driver.find_elements(By.XPATH, f"//*[@id='pcontent']/div/div[{j}]")
            for attr in attr_elements:
                elems = re.split(r"\n", attr.text)
                categories = elems[::2]
                cat_list.extend(categories)
                values = elems[1::2]
                for cat, val in zip(categories, values):
                    if cat not in all_cat_list:
                        all_cat_list.append(cat)
                    n_data[cat] = val
            
        x = [item for item in all_cat_list if item not in cat_list]
        # driver.back()
        
        n_data = pd.DataFrame([n_data])
    
        result = pd.concat([data, n_data], ignore_index=True, sort=False)
        data, result = result, data
        
        counter += 1
        now = datetime.datetime.now()
        # print(f"The number of infos entered: {counter}")
        end_time = time.time()
        diff_time = end_time - element_time
        print(f'Time for {counter}-th element: {diff_time}')
        
    end_time = time.time()
    diff_time = end_time - page_time
    print(f'Time for {n+1}-th page: {diff_time}------------------------------------------------------------------------------------------\n')
    print(f'The time at the time: {now.time()}\n')
    driver.quit()
    
end_time = time.time()
diff_time = end_time - start_time
print(f'Execution time: {diff_time}')
```
